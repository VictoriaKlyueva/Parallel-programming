```python
from matplotlib import pyplot as plt
```

```python
images = ["300px", "400px", "500px", "600px", "950px", "2400px"]
```

```python
gauss_time = {"Consistently" : [0.46, 1.12, 1.74, 2.5, 6.48, 49.2],
              "OpenMP" : [0.001, 0.001, 0.003, 0.003, 0.011, 0.078],
              "Vectorized" : [0.000, 0.001, 0.001, 0.002, 0.006, 0.041]}

mosaic_time = {"Consistently" : [0.001, 0.001, 0.003, 0.004, 0.023, 0.62],
              "OpenMP" : [0.000, 0.001, 0.001, 0.002, 0.007, 0.044],
              "Vectorized" : [0.000, 0.000, 0.001, 0.002, 0.005, 0.034]}
```

```python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(20,8))
fig.suptitle('\nPlots of time dependence on the number of pixels', fontsize=15)

# Gauss blur plot
for key in gauss_time:
  axes[0, 0].plot(images, gauss_time[key], label=key)

axes[0, 0].set_title("Gauss blur", fontsize=15)
axes[0, 0].legend()

for key in gauss_time:
  if key == "Consistently":
    continue
  axes[0, 1].plot(images, gauss_time[key], label=key)

axes[0, 1].set_title("Gauss blur paralleled", fontsize=15)
axes[0, 1].legend()


# Mosaic filter plot
for key in mosaic_time:
  axes[1, 0].plot(images, mosaic_time[key], label=key)

axes[1, 0].set_title("Mosaic filter", fontsize=15)
axes[1, 0].legend()

for key in mosaic_time:
  if key == "Consistently":
    continue
  axes[1, 1].plot(images, mosaic_time[key], label=key)

axes[1, 1].set_title("Mosaic filter paralleled", fontsize=15)
axes[1, 1].legend()
```
